```python
from arcgis import geometry #use geometry module to project Long,Lat to X and Y
from tqdm import tqdm

grid_id_list = []
input_output = dict()
for i, row in tqdm(df_burglary.iterrows(), total=df_burglary.shape[0]):
    # Get geometries iin the destination coordinate system
    input_geometry = {'y': float(row['Lat']), 'x':float(row['Long'])}
    if (input_geometry['y'], input_geometry['x']) in input_output:
        output_geometry = input_output[(input_geometry['y'], input_geometry['x'])]
    else:
        output_geometry = geometry.project(geometries = [input_geometry], in_sr = 4326,
                                      out_sr = fish_net_fset.spatial_reference["latestWkid"], gis=gis)
        input_output[(input_geometry['y'], input_geometry['x'])] = output_geometry
    xmin = -7924973.6688
    ymin = 5195182.101099998
    x_idx = (output_geometry[0].x - xmin)//300
    y_idx = (output_geometry[0].y - ymin)//300
    grid_id = (x_idx + 1) + 100 * (y_idx)
    grid_id_list.append(grid_id)
```

```python
df_burglary['gridId'] = grid_id_list
```

```python
df_burglary.to_csv("../data/burglary_with_id.csv")
```

### Aggregatino of burglaries to grid

```python
df_agg = pd.DataFrame()
```

```python
import numpy as np
df_agg['gridId'] = np.arange(1, 8501)
```

```python
df_burglary["date"] = pd.to_datetime(df_burglary["OCCURRED_ON_DATE"]).dt.date
```

```python
start_date = df_burglary["date"].sort_values().values[0]
```

```python
end_date = df_burglary["date"].sort_values().values[-1]
```

```python
from datetime import datetime, timedelta
t = np.arange(start_date, end_date + timedelta(days=1), timedelta(days=1)).astype(datetime)
```

```python
t
```

```python
type(df_burglary["date"].head(1).values[0])
```

```python
t = [a.date() for a in t]
```

```python
type(t)
```

```python
matrix = np.zeros((8500, len(t)))
```

```python
for i, row in df_burglary.iterrows():
    matrix[int(row["gridId"])][t.index(row["date"])] = 1
```

```python
for i, date in enumerate(t):
    df_agg[date] = matrix[:, i]
```
